# Part 2. Text Analysis

## 2.1 extract high requency words in description

```python
import re
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.probability import *
from itertools import chain
#from tqdm import tqdm
import codecs
```

```python
with open('stopwords.txt') as f:
    stop_words = f.read().splitlines()
stop_words = set(stop_words)
```

```python
# write your code here
# define your tokenize

tokenizer = RegexpTokenizer(r"\w+(?:[-']\w+)?") 
highrequency_words = filtered_dat['description'].str.lower().str.cat(sep='\n')
unigram_tokens = tokenizer.tokenize(highrequency_words)
stopwords = set(stop_words)
final_tokens = [w for w in unigram_tokens if w not in stop_words]
final_tokens
```

```python
# find top common words with document frequencies > 5000
# you may use function FreqDist() and sort()
words = nltk.FreqDist(final_tokens)
words
frequent_words = set([k for k, v in words.items() if v > 5000])
HighFreq = sorted(frequent_words)
HighFreq
```

```python
# save your table to 'top_common_words.txt'  
with open("HighFreq.txt", 'a') as output:
  for frequent_words in HighFreq:
    print(output.write('%s\n' % frequent_words))
```

## 2.2 Find key words for describing Shiraz using TF-IDF

```python
# select 'description' from 'variety' eqaul to  'Shiraz' 

sh=filtered_dat[filtered_dat['variety']=='shiraz']['description']
sh
```

```python
# use TfidfVectorizer to calculate TF-IDF score
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(analyzer='word', stop_words = 'english')
variable=tfidf.fit_transform(sh)
vec_idf=tfidf.idf_
tfids_result=dict(zip(tfidf.get_feature_names(),vec_idf))
tfids_result
```

```python
# find words with TF-IDF score >0.4 and sort them
tfifd=list((key,value.round(2)) for (key,value) in tfids_result.items() if value> 0.4)
tfifd.sort()
tfifd
```
